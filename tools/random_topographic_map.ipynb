```python
import pygmt
import numpy as np

def random_region(default=True):
    # use a known good region
    if default:
        return [29, 45, 16, 25]
        
    # region = [xmin, xmax, ymin, ymax]
    # xmin and xmax are the minimum and maximum longitudes 
    #   in range (-180, 180)
    # ymin and ymax are the minimum and maximum latitudes
    #   in range (-90, 90)

    # use a 16:9 aspect ratio
    width = 16
    height = 9

    xmin = np.random.randint(-180, high=180 - width)
    ymin = np.random.randint(-90, high=90 - height)

    return [xmin, xmin + width, ymin, ymin + height]
```

```python
region = random_region()
grid = pygmt.datasets.load_earth_relief(resolution="05m", region=region)
```

```python
fig = pygmt.Figure()
fig.grdcontour(grid=grid, interval=500,annotation=1000)
fig.show()
```

```python
fig.savefig("contour.eps")
```
